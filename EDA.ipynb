```python
import pandas as pd
import numpy as np
import seaborn as sns
import sklearn 
import matplotlib as mpl
import matplotlib.pyplot as plt      
import math
from collections import Counter
from sklearn.decomposition import PCA
%matplotlib inline
```

```python
def load_csv(file_name):
    df=pd.read_csv(file_name)
    return df
```

```python
def scatter_raiting_outliers(df):
    raiting_amount_cols = df["raiting_amount"]
    raiting_cols = df["raiting"]
    plt.scatter(raiting_cols,raiting_amount_cols, color= "maroon")
    plt.xlabel("Raiting")
    plt.ylabel("Raiting Amount")
    plt.show()
```

```python
def hist_eda(df):
    fig = plt.figure(figsize=(12,4))

    fig1 = fig.add_subplot(1,2,1)
    fig2 = fig.add_subplot(1,2,2)

    fig1.hist(df.raiting_amount,bins = 20, color = "lime")
    fig1.set_title("Raiting Amount Histogram")
    fig1.set_xlabel("Raiting Amount")
    fig1.set_ylabel("Frequency")

    fig2.hist(df.raiting,bins = 70, color = "chocolate")
    fig2.set_title("Raiting Histogram")
    fig2.set_xlabel("Raiting")
    fig2.set_ylabel("Frequency")

    
```

```python
def show_bars(df, name, colors):
    df_copy=df.copy()
    df_copy.sort_values(by=[name])
    df_copy[name].value_counts().plot(kind='bar', color=colors)
    plt.ylabel('Count')
    plt.xlabel(name)
    plt.title('evaluation')
```

```python
def hist_size_app(df):
    fig = plt.figure(figsize=(5, 3))
    df.size_app.hist(bins=80, density=True, rwidth=0.8, color='salmon', grid=False)
    plt.xlabel('App Size')
    plt.ylabel('Frequency')
```

```python
def pairplot_function(df):
    sns.pairplot(df, hue ='raiting')
```
