```python
import pandas as pd
import requests 
from bs4 import BeautifulSoup
from lxml import etree
import json
import numpy as np
```

```python
url = "https://apps.apple.com/us/genre/ios-games/id6014"
html = requests.get(url)
soup = BeautifulSoup(html.content, "html.parser")
links_to_pages = soup.find("ul", attrs ={"class" :"list top-level-subgenres"}) 
links_to_pages.prettify()
linkList_to_games =[]
linkList_to_pages =[]
for link in links_to_pages.find_all('a'):
    linkList_to_pages.append(link.get('href'))

for link in linkList_to_pages:
    new_html = requests.get(link)
    soup1 = BeautifulSoup(new_html.content,"html.parser")
    link_game = soup1.find("div", attrs ={"class":"grid3-column"})
    for new_link in link_game.find_all("a"):
        linkList_to_games.append(new_link.get("href"))  
      
        
```

```python
game_name_list =[]
from_age_list=[]
app_header_list =[]
raiting_by_category_list =[]
raiting_list =[]
price_list=[]
offers_list = []
raiting_amount_list =[]
seller_list= []
size_app_list =[]
age_raiting_list1=[]
age_raiting_list2=[]
age_raiting_list3=[]
age_raiting_list4=[]
age_raiting_list5=[]
age_raiting_list6=[]
```

```python



for link in  linkList_to_games:
    url_game = link
    html_game = requests.get(url_game)
    new_page = BeautifulSoup(html_game.content, "html.parser")


    try:
        game_name = etree.HTML(str(new_page))
        game_name_list.append(game_name.xpath('//h1[@class="product-header__title app-header__title"]')[0].text.strip())
    except:
        game_name_list.append(np.nan)

    try:
        from_age = new_page.find("span",{"class":"badge badge--product-title"})
        from_age_list.append(from_age.get_text().strip())
    except:
        from_age_list.append(np.nan)

    try:
        app_header = new_page.find("h2",attrs = {"class": "product-header__subtitle app-header__subtitle"})
        app_header_list.append(app_header.get_text().strip())
    except:
        app_header_list.append(np.nan)
    try:
        raiting_by_category = etree.HTML(str(new_page))
        raiting_by_category_list.append(raiting_by_category.xpath('//a[@class="inline-list__item"]')[0].text.strip())
    except:
        raiting_by_category_list.append(np.nan)
    try:    
        raiting = new_page.find("span",{"class": "we-customer-ratings__averages__display"})
        raiting_list.append(raiting.get_text().strip())
    except:
        raiting_list.append(np.nan)

    try:
        price = etree.HTML(str(new_page))
        price_list.append(price.xpath('//li[@class="inline-list_item inline-listitem--bulleted app-headerlist_item--price"]')[0].text)
    except:
        price_list.append(np.nan)
    
    try:
        offers = etree.HTML(str(new_page))
        offers_list.append(offers.xpath('//li[@class="inline-list_item inline-listitem--bulleted app-headerlist_item--in-app-purchase"]')[0].text)
    except:
        offers_list.append(np.nan)

    try:
        raiting_amount=etree.HTML(str(new_page))
        raiting_amount_list.append(raiting_amount.xpath('//p[@class="we-customer-ratings__count medium-hide"]')[0].text)
    except:
        raiting_amount_list.append(np.nan)
    try:
        seller = etree.HTML(str(new_page))
        seller_list.append(seller.xpath('//div[./dt[text()="Seller"]]//dd')[0].text.strip())
    except:
        seller_list.append(np.nan)

    try:
        size_app = etree.HTML(str(new_page))
        size_app_list.append(size_app.xpath('//div[./dt[text()="Size"]]//dd')[0].text)
    except:
        size_app_list.append(np.nan)

    age_raiting = etree.HTML(str(new_page))
    try:
        age_raiting_list1.append(age_raiting.xpath('//div[./dt[text()="Age Rating"]]//dd/span')[0].text)
    except:
        age_raiting_list1.append(np.nan)
    try:
        age_raiting_list2.append(age_raiting.xpath('//div[./dt[text()="Age Rating"]]//dd/span')[1].text)
    except:
        age_raiting_list2.append(np.nan)
    try:
        age_raiting_list3.append(age_raiting.xpath('//div[./dt[text()="Age Rating"]]//dd/span')[2].text)
    except:
        age_raiting_list3.append(np.nan)
    try:
        age_raiting_list4.append(age_raiting.xpath('//div[./dt[text()="Age Rating"]]//dd/span')[3].text)
    except:
        age_raiting_list4.append(np.nan)
    try:
        age_raiting_list5.append(age_raiting.xpath('//div[./dt[text()="Age Rating"]]//dd/span')[4].text)
    except:
        age_raiting_list5.append(np.nan)
    try:
        age_raiting_list6.append(age_raiting.xpath('//div[./dt[text()="Age Rating"]]//dd/span')[5].text)
    except:
        age_raiting_list6.append(np.nan)

df = pd.DataFrame({"game_name": game_name_list,"app_header":app_header_list,"seller":seller_list,"raiting_in_category":raiting_by_category_list,
                   "from_age":from_age_list,"raiting":raiting_list,
                   "game_price":price_list,"raiting_amount":raiting_amount_list,"offers":offers_list,
                   "size_app":size_app_list,"age_raiting1":age_raiting_list1,"age_rating2":age_raiting_list2,
                   "age_rating3":age_raiting_list3,"age_rating4":age_raiting_list4,"age_rating5":age_raiting_list5,"age_raiting6":age_raiting_list6})

df.to_csv("C:/Users/tomer/Desktop/תואר ראשון/שנה ב/file_games.csv", index = True)
```
